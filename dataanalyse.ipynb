```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("E:/forjianmo/dataEND.txt")
e = np.e
#0 v
#1 RTT
#2 T
#3 D

def ln(x):
    return np.log(x)

def f(x, y):
    v = x[0]
    R = x[1]
    T = x[2]
    D = x[3]
    n1 = y[0]
    w = y[1]
    TL = y[2]
    return n1*R+w*R*ln(v)+TL+D/v

def GradientDown():
    #0 n1  1 w 2 TL
    V = np.array([8, 0.3, 50])
    delta = 0.1
    for _ in range(0, 10000):
        dV0 = 0
        dV1 = 0
        dV2 = 0
        n = np.size(data)/4
        n = 1000
        L = []
        for _ in range(0, n):
            x = data[_]
            v = x[0]
            R = x[1]
            T = x[2]
            D = x[3]
            w = 2*(f(x, V) - T)
            #print([f(x, V), T])
            dV0 += w*R/n
            dV1 += w*R*ln(v)/n
            dV2 += w/n
            L.append((f(x, V) - T)**2)
        dV = np.sqrt(dV0**2 + dV1**2 + dV2**2)
        a = np.sort(np.array(L))
        print(np.mean(a))
        V -= delta*np.array([dV0, dV1/100, dV2*100]/dV)
        #print(np.array([dV0, dV1/100, dV2*100]/dV))
    print(V)

for x in data:
    x[0] = x[0]/8
GradientDown()

```
